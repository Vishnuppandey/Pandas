```python
data_f=pd.DataFrame({
    '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
    '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
})
data_f
```

```python
def caliqma(data_f):
    pt1=data_f['1st point']
    pt2=data_f['2nd point']
    return((pt1[0]-pt2[0])**2+(pt1[1]-pt2[1])**2)**0.5
data_f['Euclian Distance']=data_f.apply(caliqma,axis=1)
data_f
```
