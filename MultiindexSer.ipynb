```python
confirm=pd.read_csv("D:/Pandas/time_series_covid19_confirmed_global.csv")
death=pd.read_csv("D:/Pandas/time_series_covid19_deaths_global.csv")
```

```python
confirmm=confirm.melt(id_vars=['Province/State','Country/Region','Lat','Long'],var_name='Date',value_name='Confirm_case')
confirmm.head()
```

```python
deathme=death.melt(id_vars=['Province/State','Country/Region','Lat','Long'],var_name='Date',value_name='Death_case')
```

```python
new_dataf=deathme.merge(confirmm,on=['Province/State','Country/Region','Lat','Long','Date'])
new_dataf
```
